# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://drive.google.com/uc?id=1_oNqybyeIxd3lnZwD_esMTNSmqgB_hsI"

hotels_data = pd.read_csv(url)

display(hotels_data.head(), hotels_data.info(), hotels_data.describe(include='all'))
```

```python
numeric_columns = hotels_data.select_dtypes(include=['number']).columns
categorical_columns = hotels_data.select_dtypes(include=['object']).columns

hotels_data[numeric_columns] = hotels_data[numeric_columns].fillna(0)
hotels_data[categorical_columns] = hotels_data[categorical_columns].fillna("Unknown")
```

```python
hotels_data.isnull().sum()
```

# ANALISIS

```python
numeric_summary = hotels_data[numeric_columns].describe().T

plt.figure(figsize=(12, 8))
correlation_matrix = hotels_data[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Correlation Heatmap for Numeric Features")
plt.show()
```

```python
reservation_counts = hotels_data['hotel'].value_counts()
display(reservation_counts)
```

```python
filtered_data = hotels_data[hotels_data['is_canceled'] == 0].copy()

filtered_data['reservation_date'] = pd.to_datetime(
    filtered_data['arrival_date_year'].astype(str) + '-' +
    filtered_data['arrival_date_month'].astype(str) + '-' +
    filtered_data['arrival_date_day_of_month'].astype(str)
)
```

Frekuensi Reservasi berdasarkan Hotel dan Tanggal

```python
hotel_date_counts = filtered_data.groupby(['hotel', 'reservation_date']).size().reset_index(name='reservation_count')
hotel_date_counts = hotel_date_counts.sort_values(by='reservation_count', ascending=False)
print("Top 10 hotel dengan frekuensi reservasi terbanyak:")
display(hotel_date_counts.head(10))

plt.figure(figsize=(10, 6))
sns.lineplot(data=hotel_date_counts, x='reservation_date', y='reservation_count', hue='hotel')
plt.title("Frekuensi Reservasi per Hotel dan Tanggal")
plt.xlabel("Tanggal Reservasi")
plt.ylabel("Jumlah Reservasi")
plt.xticks(rotation=45)
plt.show()
```

Segmentasi Pasar berdasarkan Saluran Distribusi

```python
market_segment_counts = hotels_data.groupby('market_segment')['hotel'].value_counts().unstack().fillna(0)
print("\nSegmentasi Pasar berdasarkan Saluran Distribusi:")
display(market_segment_counts)

market_segment_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title("Segmentasi Pasar berdasarkan Saluran Distribusi")
plt.xlabel("Segment Pasar")
plt.ylabel("Jumlah Reservasi")
plt.xticks(rotation=45)
plt.show()
```

Perbandingan tamu berulang vs tamu baru

```python
repeated_guest_rate = hotels_data['is_repeated_guest'].value_counts(normalize=True)
print("\nPerbandingan Tamu Berulang dan Tamu Baru secara Keseluruhan:")
display(repeated_guest_rate)

plt.figure(figsize=(6, 4))
sns.barplot(x=repeated_guest_rate.index, y=repeated_guest_rate.values)
plt.title("Perbandingan Tamu Berulang dan Tamu Baru (Keseluruhan)")
plt.xlabel("Tamu Berulang vs Baru")
plt.ylabel("Proporsi")
plt.show()

# Analisis tamu berulang dan baru berdasarkan hotel
hotel_guest_rate = hotels_data.groupby(['hotel', 'is_repeated_guest']).size().reset_index(name='count')
hotel_guest_rate = hotel_guest_rate.pivot(index='hotel', columns='is_repeated_guest', values='count').fillna(0)
hotel_guest_rate['total_guests'] = hotel_guest_rate.sum(axis=1)
hotel_guest_rate['repeated_guest_ratio'] = hotel_guest_rate[1] / hotel_guest_rate['total_guests']
hotel_guest_rate['new_guest_ratio'] = hotel_guest_rate[0] / hotel_guest_rate['total_guests']

print("\nPerbandingan Tamu Berulang dan Baru Berdasarkan Hotel:")
display(hotel_guest_rate)
```
